#required classes

```python
class Card():
    def __init__(self,rank,suit):
        self.rank = rank
        self.suit = suit
        self.value = rank_value[rank]
    def display(self):
        print("{} of {}".format(self.rank,self.suit))     
    def givevalue(self):
        return self.value
    def __str__(self):
        return "{} of {}".format(self.rank,self.suit)
    def __len__(self):
        return self.value
```

```python
class Deck():
    def __init__(self):
        self.cards=[]
        for rank in ranks:
            for suit in suits:
                obj=Card(rank,suit)
                self.cards.append(obj)
    def shuffle_deck(self):
        random.shuffle(self.cards)
    def deal_top_card(self):
        return self.cards.pop(0)
    def add_cards(self,obj):
        if type(obj)==list:
            self.cards.extend(obj)
        else:
            self.cards.append(obj)    
```

```python
class Player():
    def __init__(self,name,balance):
        self.name=name
        self.reveal=True
        self.balance =balance
        self.hand=[]
    def take_cards(self,deck,num):
        for i in range(0,num):
            self.hand.append(deck.deal_top_card())
    def show_hand(self):
        for each in self.hand:
            print(each)
    def add_money(self,money):
        self.balance+=money
    def remove_money(self,money):
        if self.balance-money<=0:
            return False
        else:
            self.balance-=money
            return True
    def discard_cards(self):
        self.hand=[]
        
```

```python
class Banker():
    def __init__(self,balance):
        self.name="banker"
        self.balance=balance
        self.hand=[]
        self.reveal=False
    def take_cards(self,deck,num):
        for i in range(0,num):
            self.hand.append(deck.deal_top_card())
    def show_hand(self):
        for each in self.hand:
            print(each)
    def add_money(self,money):
        self.balance+=money
    def remove_money(self,money):
        if self.balance-money<=0:
            return False
        else:
            self.balance-=money
            return True
    def discard_cards(self):
        self.hand=[]
```

#required functions

```python
def disp(player_obj):
    if player_obj.reveal==False:
        for each in player_obj.hand[::]:
            print("-----------")
            print("|         |")
            print("|    **   |")
            print("|    **   |")
            print("|         |")
            print("-----------")
        
    else:
        for each in player_obj.hand[::]:
            print("-----------")
            print("|         |")
            print(f"  {each.rank}  ")
            print(f"  {each.suit}  ")
            print("|         |")
            print("-----------")

def disp_banker(player_obj):
    for each in player_obj.hand[::]:
        print("-----------")
        print("|         |")
        print(f"  {each.rank}  ")
        print(f"  {each.suit}  ")
        print("|         |")
        print("-----------")
```

```python
def welcome_message():
    print("Welcome to the game. We will now begin")
    while True:
        print(".....................................")
        print(".....................................")
        print(".....................................")
        ip=input("Press Enter to proceed:")
        if ip=="":
            break
def exit_message():
    print("thank you for playing the game")
    while True:
        print(".....................................")
        print(".....................................")
        print(".....................................")
        ip=input("Press Enter to proceed:")
        if ip=="":
            break
    
```

```python
def replay():
    while True:
        ip=input("Press Y/y if you want to replay, else press N/n :")
        if ip in ["Y","y","N","n"]:
            break
        else:
            print("Enter valid value")
    if ip in  ["Y","y"]:
        return True
    else:
        return False
```

```python
def player_input():
    name=input("Enter player name: ")
    while True:
        balance=input("Enter money you want to begin with: ")
        if balance.isdigit()==True:
            break
        else:
            print("...Enter valid money value...")
    return name,int(balance)
    
```

```python
def banker_input():
    while True:
        balance=input("Enter money you want the banker to begin with: ")
        if balance.isdigit()==True:
            break
        else:
            print("...Enter valid money value...")
    return int(balance)
```

```python
def buffer():
    k=input("<<<<<<<<<<press Enter key to proceed>>>>>>>>>")
```

```python
def game_number():
    while True:
        num=input("Enter a number between 3-5, both you and the banker will be dealt that number of cards: ")
        if num.isdigit()==True and int(num) in range(3,6):
            break
        else:
            print("Enter a valid number between 3 and 5")
    return int(num)
```

```python
def deal_done(num):
    print(f"{num} cards have been dealt to you and the banker: ")
```

```python
def place_bet(player_obj):
    while True:
        num=input("Enter the bet money(minimum value of bet is 100): ")
        if num.isdigit()==True and int(num) in range(100,player_obj.balance+1):
            break
        else:
            print(f"Enter a valid number between 100 and {player_obj.balance}")
    return int(num)
```

```python
def compare_hand(player_obj,banker_obj):
    sum_p=0
    sum_b=0
    for each in player_obj.hand:
        sum_p+=each.value
    for each in banker_obj.hand:
        sum_b+=each.value
    if sum_p>=sum_b:
        return True
    else:
        return False
```

```python
def player_win(player_obj,banker_obj,bet):
    global win
    player_obj.add_money(bet)
    if banker_obj.remove_money(bet):
        print(f"player balance: {player_obj.balance}")
        print(f"banker balance: {banker_obj.balance}")
    else:
        print(f"player balance: {player_obj.balance}")
        win=1
        print(f"banker is bankrupt")
def banker_win(player_obj,banker_obj,bet):
    global win
    banker_obj.add_money(bet)
    if player_obj.remove_money(bet):
        print(f"player balance: {player_obj.balance}")
        print(f"banker balance: {banker_obj.balance}")
    else:
        print(f"banker balance: {banker_obj.balance}")
        win=2
        print(f"player is bankrupt")
```

```python
def quit_check():
    while True:
        ip=input("Press Y/y if you want to QUIT, else press N/n :")
        if ip in ["Y","y","N","n"]:
            break
        else:
            print("Enter valid value: ")
    if ip in  ["Y","y"]:
        return True
    else:
        return False
```

```python
def reset():
    global win
    win =0
```

#game mechanics

```python
while True:
    welcome_message()
    deck=Deck()
    deck.shuffle_deck()
    b1=banker_input()
    banker=Banker(b1)
    n,b=player_input()
    player=Player(n,b)
    num=game_number()
    while True:
        print("cards will now be dealt: ")
        buffer()
        player.take_cards(deck,num)
        banker.take_cards(deck,num)
        deal_done(num)
        buffer()
        print("Your hand: ")
        disp(player)
        buffer()
        print("Banker's hand: ")
        disp(banker)
        buffer()
        bet=place_bet(player)
        print("Your hand: ")
        disp(player)
        buffer()
        print("Banker's hand: ")
        disp_banker(banker)
        buffer()
        if compare_hand(player,banker):
            player_win(player,banker,bet)
        else:
            banker_win(player,banker,bet)
        buffer()
        if win==0:
            player.discard_cards()
            banker.discard_cards()
            print("next round will begin: ")
            buffer()
            if quit_check():
                break
            else:
                buffer()
            
        elif win==1:
            player.discard_cards()
            banker.discard_cards()
            print(f"player has won with a balance of {player.balance}")
            break
        else:
            player.discard_cards()
            banker.discard_cards()
            print(f"banker has won with a balance of {banker.balance}")
            break
    if win==0:
        print("you have chosen to quit current game")
    elif win==1:
        print("you have beaten the banker")
    elif win==2:
        print("the banker has beaten you")
    else:
        pass
        
    if replay()==False:
        exit_message()
        break    
    else:
        reset()
        del player
        del banker
```
